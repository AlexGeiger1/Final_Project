```python
#Importing Dependencies
import pandas as pd
from datetime import date
import random
```

```python
#Reading in CSV file
raw_csv= "../US_econ_stats.csv"

econ_df = pd.read_csv(raw_csv)
```

```python
econ_df
```

```python
mapping = {econ_df.columns[0]: 'Stats'}
econ_df = econ_df.rename(columns=mapping)
econ_df
```

```python
econ_df.dtypes
```

```python
econ_df['Stats'] = econ_df['Stats'].astype('string')
econ_df.dtypes
```

```python
econ_df = econ_df.dropna()
```

```python
econ_df['Stats'].tolist()
```

```python
stats_wanted = ["Real_GDP_growth (Annual growth %)", "Net_saving_rate_in_household_disposable_income (%)", 
                "Tax_on_corporate_profits (% of GDP)", "Taxes_on_goods_and_services (% of GDP)", 
                "Imports_of_goods_and_services (% of GDP)", "Exports_of_goods_and_services (% of GDP)", 
                "Inflation_rate:_all_items (Annual growth %)", "Long-term_interest_rates (%)", 
                "Unemployment_rate:_total_labour_force (%)", "Tertiary_attainment_in_population_aged_25-64 (%)",
                "Life_expectancy_at_birth (Years)", "Population_growth_rates (%)", 
                "Purchasing_power_parities (USD per USD)", 
                "Labour_compensation_per_unit_labour_input,_total_economy (Annual growth %)"]

new_econ_df = econ_df[econ_df.Stats.isin(stats_wanted)]
new_econ_df
```

```python
indx_econ_df = new_econ_df.set_index("Stats")
indx_econ_df
```

```python
tp_econ_df = indx_econ_df.T
tp_econ_df
```

```python
tp_econ_df.dtypes
```

```python
tp_econ_df.reset_index(inplace=True)
tp_econ_df
```

```python
mapping = {tp_econ_df.columns[0]: 'Year'}
tp_econ_df = tp_econ_df.rename(columns=mapping)
tp_econ_df
```

```python
tp_econ_df = tp_econ_df.rename_axis(None, axis = 1)
```

```python
tp_econ_df
```

```python
tp_econ_df.rename(columns = {"Real_GDP_growth (Annual growth %)": "Ann_real_GDP_gwth_pct", 
                "Net_saving_rate_in_household_disposable_income (%)": "Net_savings_rate_household_disp_inc_pct", 
                "Tax_on_corporate_profits (% of GDP)": "Tax_on_corp_profits_PctofGDP",
                "Taxes_on_goods_and_services (% of GDP)": "Tax_on_goods_and_services_PctofGDP", 
                "Imports_of_goods_and_services (% of GDP)": "Imports_of_goods_and_services_PctofGDP",
                "Exports_of_goods_and_services (% of GDP)": "Exports_of_goods_and_services_PctofGDP", 
                "Inflation_rate:_all_items (Annual growth %)": "Infl_rate_all_items_ann_grwth_pct",
                "Long-term_interest_rates (%)": "LT_int_rate_pct", 
                "Unemployment_rate:_total_labour_force (%)": "Unemp_rate_total_pct",
                "Tertiary_attainment_in_population_aged_25-64 (%)": "Teritary_attainment_aged_25_to_64_pct",
                "Life_expectancy_at_birth (Years)": "Life_expectancy_yrs",
                "Population_growth_rates (%)": "Pop_grwth_rate_pct", 
                "Purchasing_power_parities (USD per USD)": "PPI_USD_per_USD", 
                "Labour_compensation_per_unit_labour_input,_total_economy (Annual growth %)": "Labor_comp_per_unit_labor_input_ann_growth_pct"
}, inplace = True)
```

```python
tp_econ_df
```

```python
tp_econ_df.dtypes
```

```python
tp_econ_df['Year'] = pd.to_datetime(tp_econ_df['Year'], format='%Y')
tp_econ_df['Year'] = tp_econ_df['Year'].dt.to_period('Y')
```

```python
tp_econ_df
```

```python
tp_econ_df.dtypes
```

```python
tp_econ_df.to_csv("final_econ_data.csv", index=False)
```
